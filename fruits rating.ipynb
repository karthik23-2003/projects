```python
import pandas as pd
```

```python
fruits = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Fruits.csv')
```

```python
fruits.head()
```

```python
fruits.info()
```

```python
fruits.describe()
```

```python
 fruits.columns
```

```python
y = fruits['Fruit Colour Score']
x = fruits[['Fruit Category', 'Fruit Weight', 'Fruit Width','Fruit Length']]
```

```python
fruits.shape
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
